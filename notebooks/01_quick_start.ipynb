## 3. Run a Pipeline Programmatically

```python
from vision_workbench.core.config import PipelineConfig, DataStageConfig, RuntimeConfig
from vision_workbench.core.context import PipelineContext
from vision_workbench.core.orchestrator import PipelineOrchestrator
from vision_workbench.pipeline.data.stage import DataStage
from pathlib import Path

# Build a simple data pipeline
config = PipelineConfig(
    name="notebook-test",
    stages=["data"],
    runtime=RuntimeConfig(workspace=Path("./vw_workspace")),
    data=DataStageConfig(
        enabled=True,
        source="./tests/fixtures/mini_coco/images/train/",
        target="./vw_workspace/data/processed/test-ds/",
        dataset_name="test-ds",
    ),
)

orch = PipelineOrchestrator(config)
orch.register_stage(DataStage())

# Dry run first
plan = orch.dry_run()
for step in plan:
    print(f"[{step['stage']}] {step['summary']}")
```
